```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_titles.csv", encoding='iso-8859-1')

data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
movies = data.drop(columns=[col for col in data.columns if "Unnamed" in col])
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.shape
```

```python
movies['tags'] = (movies['director'] + ' ' + 
                  movies['cast'] + ' ' + 
                  movies['listed_in'] + ' ' + 
                  movies['description'])
```

```python
movies.head()
```

```python
movies = movies[['show_id', 'type', 'title', 'tags']]

movies
```

```python
movies.dropna(inplace=True)
```

```python
movies = movies.dropna()
```

```python
movies
```

```python
movies.reset_index(inplace=True, drop=True)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5702, stop_words='english')
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vector = cv.fit_transform(movies['tags'].values.astype('U')).toarray()
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
movies[movies['title']=='Zubaan'].index[0]
```

```python
similarity[5699]
```

```python
similar_movies = sorted(list(enumerate(similarity[5699])), reverse=True, key= lambda vector: vector[1])
```

```python
for i in similar_movies[1:6]: 
    print(movies.iloc[i[0]].title)
    print(i[1])
```

```python
def recommend_movie(movie_name): 
    movie_index = movies[movies['title'] == movie_name].index[0]
    similar_movies = sorted(list(enumerate(similarity[movie_index])), reverse=True, key= lambda vector: vector[1])
    for i in similar_movies[1:6]:
        similarity_percentage = round(i[1] * 100, 2)
        print(f'{movies.iloc[i[0]].title} is similar to {movie_name} with a similarity of {similarity_percentage}%.')
```

```python
recommend_movie('Midnight Mass')
```
